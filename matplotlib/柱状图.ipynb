```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.arange(5)
lr = [31,31,15,11,12]
rf = [40,23,16,9,12]
ct = [55,16,8,8,14]
```

```python
mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False

tol_width,n = 0.8,3
width = tol_width/n
x = x-(tol_width - width)/2
plt.bar(x,lr,width=width,label="逻辑回归")
plt.bar(x+width,rf,width=width,label="随机森林", 
        tick_label=['极低','低','中等','高','极高'])
plt.bar(x+2*width,ct,width=width,label="CatBoost")
plt.grid(True, axis="y", ls=":", color="c", alpha=0.3)
plt.ylabel("占总面积的百分比(%)")
plt.xlabel("滑坡敏感性等级划分")
plt.legend()
plt.tight_layout()
plt.savefig("等级划分柱状图.jpg",dpi=300)
plt.show()
```
