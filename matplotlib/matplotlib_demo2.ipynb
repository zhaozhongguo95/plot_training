```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from matplotlib import cm
```

### demo_1

```python
fig = plt.figure(figsize=(10,8),dpi=300)
mpl.rcParams["font.sans-serif"]=["SimHei"]
mpl.rcParams["axes.unicode_minus"]=False
####柱状图
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [3, 1, 4, 5, 8, 9, 7, 2]
ax1 = fig.add_subplot(221,xlabel="箱子编号",ylabel="箱子重量（kg）",title="A")
ax1.grid(True, axis="y", ls=":", color="r", alpha=0.3)
ax1.bar(x,y,align="center",color = "b",
       tick_label=['q','a','c','e','r','j','b','p'],hatch='/')

#####条形图
ax2 = fig.add_subplot(222,xlabel="箱子编号",ylabel="箱子重量（kg）")
ax2.barh(x,y,align="center",color = "c",
       tick_label=['q','a','c','e','r','j','b','p'],hatch='/')

#####直方图
mpl.rcParams.update({'font.size': 14})
ax3 = fig.add_subplot(223,xlabel="箱子重量（kg）",ylabel="销售数量（个）")
# 设置测试成绩
boxWeight = np.random.randint(0, 10, 100)
bins = range(0, 11, 1)
ax3.hist(boxWeight,bins=bins,color="b",histtype="bar",rwidth=1,alpha=0.6)


####饼图
mpl.rcParams.update({'font.size': 16})
ax4 = fig.add_subplot(224)

kinds = "简易箱", "保温箱", "行李箱", "密封箱"
colors = ['b','g','r','c'] 
soldNums = [0.05, 0.45, 0.15, 0.35]
ax4.pie(soldNums,labels=kinds,autopct="%3.1f%%",startangle=60,colors=colors)

plt.show()
```

```python
####

fig= plt.figure(figsize=(8,7))
#左下宽高
fig_ax1 = fig.add_axes([0.1,0.24,0.5,0.5])
fig_ax1.set_title("(a)",loc='left')
fig_ax1.scatter(x,y)
fig_ax1.axhspan(ymin=2, ymax=6, facecolor="c", alpha=0.3)

fig_ax2 = fig.add_axes([0.7,0.24,0.5,0.5])
fig_ax2.set_title("(b)",loc='left')
fig_ax2.set_xlabel("x")
fig_ax2.set_ylabel("y")
fig_ax2.scatter(x,y,c="r")
fig_ax2.axvline(x=2, c="b",ls="--",lw=2)
fig_ax2.plot(x,y,'c')
fig_ax2.legend("2sin(x)",loc="lower right")
fig_ax2.set_title("sin(x) 曲线")
plt.text(5, 6, "y=sin(x)", weight="bold", color="black")

plt.show()
```

### 极线图

```python
###极线图
barSlices = 12
theta = np.linspace(0.0, 2*np.pi, barSlices, endpoint=False)
r = 30*np.random.rand(barSlices)


plt.polar(theta, r, color="chartreuse", linewidth=2,
          marker="*", mfc="b", ms=10)

plt.show()
```

```python
####气泡图
fig= plt.figure(figsize=(12,10),dpi=300)
a = np.random.randn(100)
b = np.random.randn(100)

fig_ax1 = fig.add_axes([0.1,0.24,0.5,0.5])
fig_ax1.set_title("(a)",loc='left')
fig_ax1.scatter(a, b, s=np.power(10*a+20*b, 2), c=np.random.rand(100),
            cmap=mpl.cm.RdYlBu, marker="o")

###棉花棒
a = np.linspace(0.5, 2*np.pi, 20)
b = np.random.randn(20)

fig_ax2 = fig.add_axes([0.68,0.24,0.5,0.5])
fig_ax2.set_title("(b)",loc='left')
fig_ax2.stem(a, b, linefmt="-.", markerfmt="o", basefmt="-")

####箱线图
fig_ax3 = fig.add_axes([0.1,-0.33,0.5,0.5])
fig_ax3.set_title("(c)",loc='left')

mpl.rcParams["font.sans-serif"] = ["FangSong"]
mpl.rcParams["axes.unicode_minus"] = False

x = np.random.randn(1000)
fig_ax3.boxplot(x)
fig_ax3.set_xticks([1], ["随机数生成器AlphaRM"])
fig_ax3.set_ylabel("随机数值")
fig_ax3.set_title("随机数生成器抗干扰能力的稳定性")
fig_ax3.grid(axis="y", ls=":", lw=1, color="gray", alpha=0.4)

###误差图
fig_ax4 = fig.add_axes([0.68,-0.33,0.5,0.5])
fig_ax4.set_title("(D)",loc='left')

x = np.linspace(0.1, 0.6, 6)
y = np.exp(x)
#yerr：y轴方向的数据点的误差计算方法
#xerr：x轴方向的数据点的误差计算方法
fig_ax4.errorbar(x, y, fmt="bo:", yerr=0.2, xerr=0.02)

fig_ax4.set_xlim(0, 0.7)
plt.show()
```

```python
fig= plt.figure(figsize=(12,10),dpi=300)
####堆积柱状图
x = [1, 2, 3, 4, 5]
y = [6, 10, 4, 5, 1]
y1 = [2, 6, 3, 8, 5]
#左下宽高
fig_ax1 = fig.add_axes([0.1,0.24,0.5,0.5])
fig_ax1.set_title("(a)",loc='left')
fig_ax1.bar(x, y, align="center", color="#66c2a5", tick_label=["A", "B", "C", "D", "E"], label="班级A")
fig_ax1.bar(x, y1, align="center", bottom=y, color="#8da0cb", label="班级B")
fig_ax1.set_xlabel("测试难度")
fig_ax1.set_ylabel("试卷份数")
fig_ax1.legend()


####条状柱状图
fig_ax2 = fig.add_axes([0.68,0.24,0.5,0.5])
fig_ax2.set_title("(b)",loc='left')
fig_ax2.barh(x, y, align="center", color="#66c2a5", tick_label=["A", "B", "C", "D", "E"], label="班级A")
fig_ax2.barh(x, y1, align="center", left=y, color="#8da0cb", label="班级B")
fig_ax2.set_xlabel("测试难度")
fig_ax2.set_ylabel("试卷份数")
fig_ax2.grid(True, axis="x", ls=":", color="r", alpha=0.3)


####并列柱状图
fig_ax3 = fig.add_axes([0.1,-0.33,0.5,0.5])
fig_ax3.set_title("(c)",loc='left')

x = np.arange(5)
y = [6, 10, 4, 5, 1]
y1 = [2, 6, 3, 8, 5]
bar_width = 0.35
tick_label = ["A", "B", "C", "D", "E"]

fig_ax3.bar(tick_label, y, bar_width, align="center", color="c", label="班级A", alpha=0.5)
fig_ax3.bar(x+bar_width, y1, bar_width, color="b", align="center", label="班级B", alpha=0.5)
fig_ax3.set_xlabel("测试难度")
fig_ax3.set_ylabel("试卷份数")
fig_ax3.set_xticks(x+bar_width/2, tick_label)
fig_ax3.legend()

#####
fig_ax4 = fig.add_axes([0.68,-0.33,0.5,0.5])
fig_ax4.set_title("(D)",loc='left')
x = np.arange(1, 6, 1)
y = [6, 10, 4, 5, 1]
y1 = [2, 6, 3, 8, 5]
y2 = [3, 4, 5, 6, 2]
labels = ["BluePlanet", "BrownPlanet", "GreenPlanet"]
colors = ["#8da0cb", "#fc8d62", "#66c2a5"]

fig_ax4.stackplot(x, y, y1, y2, labels=labels, colors=colors)
fig_ax4.legend(loc="upper left")

plt.show()
```

```python
####
fig= plt.figure(figsize=(12,10),dpi=300)

fig_ax1 = fig.add_axes([0.1,0.24,0.5,0.5])
fig_ax1.set_title("(a)",loc='left')
fig_ax2 = fig.add_axes([0.68,0.24,0.5,0.5])
fig_ax2.set_title("(b)",loc='left')
fig_ax3 = fig.add_axes([0.1,-0.35,0.5,0.5])
fig_ax3.set_title("(c)",loc='left')
fig_ax4 = fig.add_axes([0.68,-0.35,0.5,0.5])
fig_ax4.set_title("(D)",loc='left')
####
fig_ax1.broken_barh([(30, 100), (180, 50), (260, 60)], (20, 8), facecolors="#1f78b4")
fig_ax1.broken_barh([(60, 90), (190, 20), (230, 30), (280, 60)], (10, 8), facecolors=(
        "#7fc97f", "#beaed4", "#fdc086", "#ffff99"))
fig_ax1.set_xlim(0, 360)
fig_ax1.set_ylim(5, 35)
fig_ax1.set_xlabel("演出时间")
fig_ax1.set_xticks(np.arange(0, 361, 60))
fig_ax1.set_yticks([15,25],["歌剧院A", "歌剧院B"])
fig_ax1.grid(ls="-.", lw=1, color="gray")
fig_ax1.set_title("不同地区的歌剧院的演出时间比较")

####阶梯图
x = np.linspace(1, 10, 10)
y = np.sin(x)
fig_ax2.step(x, y, color="#8dd3c7", where="pre", lw=2)
fig_ax2.set_xlim(0, 11)
fig_ax2.set_xticks(np.arange(1, 11, 1))
fig_ax2.set_ylim(-1.2, 1.2)

####堆积直方图
mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
scoreT1 = np.random.randint(0, 100, 100)
scoreT2 = np.random.randint(0, 100, 100)
x = [scoreT1, scoreT2]
colors = ["#8dd3c7", "#bebada"]
labels = ["班级A", "班级B"]
bins = range(0, 101, 10)
#histtype:["stepfilled","bar"]
fig_ax3.hist(x, bins=bins, color=colors, histtype="bar", rwidth=10,
         stacked=True, label=labels)
fig_ax3.set_xlabel("测试成绩")
fig_ax3.set_ylabel("学生人数")
fig_ax3.set_title("不同班级的测试成绩的直方图")
fig_ax3.legend(loc="upper left")

####饼图
mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
labels = "A难度水平", "B难度水平", "C难度水平", "D难度水平"
students = [0.35, 0.15, 0.20, 0.30]
colors = ["#377eb8", "#4daf4a", "#984ea3", "#ff7f00"]         
explode = (0.1, 0.1, 0.1, 0.1)
#explode控制是否分裂，
fig_ax4.pie(students, explode=explode, labels=labels, autopct="%3.1f%%",
        startangle=45, shadow=True, colors=colors)

fig_ax4.set_title("选择不同难度测试试卷的学生百分比")
fig_ax4.legend(loc="lower left",fontsize=12, bbox_to_anchor=(0.91, 0, 0.3, 1))

plt.show()
```

```python
####内嵌环形饼图
fig= plt.figure(figsize=(6,8),dpi=100)
elements = ["面粉", "砂糖", "奶油", "草莓酱", "坚果"]
weight1 = [40, 15, 20, 10, 15]
weight2 = [30, 25, 20, 15, 10]
colormapList = ["#e41a1c", "#377eb8", "#4daf4a", "#984ea3", "#ff7f00"]
outer_colors = colormapList;inner_colors = colormapList
wedges1, texts1, autotexts1 = plt.pie(weight1, autopct="%3.1f%%",radius = 1,pctdistance=0.85,
                                      colors=outer_colors,textprops=dict(color="w"),
                                      wedgeprops=dict(width=0.3, edgecolor="w"))

wedges2, texts2, autotexts2 = plt.pie(weight2,autopct="%3.1f%%", radius = 0.7,  pctdistance=0.75,
                                      colors=outer_colors, textprops=dict(color="w"),
                                      wedgeprops=dict(width=0.3, edgecolor="w"))
plt.setp(autotexts1, size=15, weight="bold")
plt.setp(autotexts2, size=15, weight="bold")
plt.setp(texts1, size=12)
plt.title("不同果酱面包配料比例表的比较",fontsize=14)
plt.show()
```
