```python
import matplotlib.pylab as plt
import matplotlib as mpl
import numpy as np
```

```python
x = np.linspace(0.05,10,1000)
y = np.sin(x)

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
plt.rcParams.update({'font.size':20})
fig = plt.figure(figsize=(16,12))
ax1 = fig.add_subplot(221)
ax1.plot(x,y,ls="-.",lw=2,c="black",label="sin(x)")
ax1.set_title("sin(x) 曲线")

plt.xlim(0.05,10)
plt.ylim(-2,2)

plt.legend(loc="lower right")

plt.axhline(y=0, c="g",ls="--",lw=2)
plt.axvline(x=2, c="b",ls="--",lw=2)

plt.xlabel("x轴")
plt.ylabel("y轴")

plt.grid(linestyle=":", color="r")

plt.axhspan(ymin=0.0, ymax=0.5, facecolor="c", alpha=0.3)
plt.text(3.10, 0.09, "y=sin(x)", weight="bold", color="b")
plt.show()
```
